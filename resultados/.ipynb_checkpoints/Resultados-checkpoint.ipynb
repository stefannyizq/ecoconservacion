### 1. Propósito del AHP

El Proceso Analítico Jerárquico (AHP) se utilizó para asignar pesos relativos a los criterios considerados en el análisis de idoneidad territorial. Este enfoque permite integrar múltiples capas temáticas como pendiente, uso del suelo o vías en una sola medida agregada: el Índice de Idoneidad (II). 

Los criterios evaluados fueron:

- Pendiente  
- Uso del suelo  
- Curvas de nivel  
- Vías  
- Áreas protegidas  
- Hidrografía

### 2. Matriz de comparación por pares

Se construyó una matriz de comparación 6x6 basada en la escala de Saaty (1 a 9), donde cada valor representa la importancia relativa de un criterio sobre otro.

Por ejemplo:
- La **pendiente** fue considerada 3 veces más importante que el uso del suelo (valor 3).
- Los valores son recíprocos: si A es 3 veces más importante que B, entonces B tiene un valor de 1/3 respecto a A.


### 3. Pesos normalizados por criterio

### 4. Validación de consistencia

Para asegurar que los juicios de comparación no sean arbitrarios ni contradictorios, se calcularon los siguientes indicadores:

Este índice representa el valor relativo de cada celda espacial (unidad de análisis) para aportar a la conservación de la biodiversidad y la conectividad ecológica en el territorio agrícola, alineándose con el objetivo central del proyecto.

## Metodología de Priorización Espacial

Con el fin de enfocar las acciones de conservación y planificación en los sectores más estratégicos, se implementa un enfoque cuantitativo basado en el **percentil 75** del Índice de Idoneidad. Este umbral ha sido seleccionado para identificar el **cuartil superior** de las celdas, es decir, aquellas con el 25% de los valores más altos de idoneidad, bajo la hipótesis de que estas áreas concentran las condiciones más favorables para la provisión de servicios ecosistémicos y la conectividad biológica.

**El procedimiento incluye:**

  Se determina el valor umbral que delimita el 25% superior de las celdas evaluadas, asegurando que solo se consideren aquellas con condiciones ecológicas sobresalientes.

- **Filtrado espacial:**
